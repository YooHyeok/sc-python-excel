```python
import xlwings as xw # xlwinngs import 및 xw 별칭 부여

# 액셀 앱 만들기
app = xw.App(add_book=False) # 액셀 생성 및 오픈
app # 출력
```

```python
# 액셀 워크북 만들기
wb = app.books.add() # 생성되어 열린 액셀파일에 워크시트(통합문서 2) 생성
wb # 출력
```

```python
# 워크시트 선택(이름)
ws = wb.sheets['Sheet1']
ws
```

```python
# 시트 이름 변경
ws.name = "영업1팀"
```

```python
# 다른 이름으로 액셀 저장
wb.save('교육이슈현황.xlsx')
```

```python
# 워크북 닫기
wb.close()
```

```python
# 액셀 앱 닫기
app.quit()
```
